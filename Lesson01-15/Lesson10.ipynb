```python
count = 0
with open("jinko.txt") as f:
    l = f.readline()
    while l:
        l = f.readline()
        m = l.strip()
        print(m)
        count = count + 1
        if count >= 5:
            break
```

```python
with open("jinko.txt") as f:
    l = f.readline()
    while l:
        l = f.readline()
        v = l.split(",")
        print(v)
```

```python
import csv

with open("jinko.txt") as f:
    reader = csv.reader(f)
    for l in reader:
        print(l)
```

```python
with open("jinko.txt") as f:
    reader = csv.reader(f)
    for l in reader:
        if l[0] == "北海道":
            print(l)
```

```python
s = 0
with open("jinko.txt") as f:
    reader = csv.reader(f)
    for l in reader:
        if l[0] == "北海道":
            s = s + int(l[2])

print("北海道の人口", s)
```

```python
with open("jinko_sjis.txt") as f:
    while True:
        l = f.readline()
        print(l)
        if not l:
            break
```

```python
with open("jinko_sjis.txt", encoding='shift-jis') as f:
    while True:
        l = f.readline()
        print(l)
        if not l:
            break
```

```python
with open("jinko_sjis.txt", encoding='shift-jis', newline='\r\n') as f:
    while True:
        l = f.readline()
        print(l)
        if not l:
            break
```

```python
data = [1, 2, 3]

for i in range(len(data)):
    data[i] = str(data[i])

csvdata = ",".join(data)
print(csvdata)
```

```python
alldata = [
   [1, 2, 3],
   [2, 3, 5]
]
with open("writeexample3.txt", "w") as f:
        for data in alldata:
            csvdata = ",".join([str(x) for x in data])
            f.write(csvdata + "\r\n")
```

```python
import csv

alldata = [
   [1, 2, 3],
   [2, 3, 5]
]

with open("writeexample4.txt", "w") as f:
    writer = csv.writer(f)
    writer.writerows(alldata)
```

```python
import csv

alldata = [
   [1, 2, 3],
   [2, 3, 5]
]

with open("writeexample5.txt", "w") as f:
    writer = csv.writer(f)
    for data in alldata:
            writer.writerow(data)
```
