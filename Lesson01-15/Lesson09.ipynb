```python
class Dog:
    # property
    voice = "bow!"

    # constructor
    def __init__(self, name = "Buddy"):
        self.name = name

    # method
    def bark(self):
        print(self.voice)

    # class method
    @classmethod
    def description(self):
        print(f"犬の鳴き声は{self.voice}")
```

```python
Dog.description()

buddy = Dog()
print(buddy.name)
buddy.bark()

jake = Dog("Jake")
print(jake.name)
jake.bark()
```
