```python
# ライブラリのインポートなど
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# 画像の読み込み
saya_org = Image.open("saya160104294609_TP_V4.jpg")

# 画像の表示
plt.imshow(saya_org)
```

```python
saya_org_array = np.array(saya_org)
```

```python
saya_org.width
```

```python
saya_org.height
```

```python
saya_org.size
```

```python
# 土台の作成
#（new()の最初の引数は色空間の指定、2つ目の引数は(幅, 高さ)のタプル）
saya2 = Image.new("RGB", (saya_org.width * 2, saya_org.height))

# 画像の貼り付け
#（paste()の最初の引数は貼り付ける画像、2つ目の引数は左上の座標(横, 縦)のタプル）
saya2.paste(saya_org, (0, 0))
saya2.paste(saya_org, (saya_org.width, 0))

# 新しい画像の表示
plt.imshow(saya2)
```

```python
# 顔の部分の切り抜き
saya_crop = saya_org.crop((300, 100, 500, 250))

# 切り抜いた画像の表示
plt.imshow(saya_crop)
```

```python
# 画像を横4枚に分割
saya_crop_list = []
for i in range(4):
    saya_crop_list.append(saya_org.crop((200 * i, 0, 200 * (i + 1), saya_org.height)))

# 左端の画像のみ表示
plt.imshow(saya_crop_list[0])
```

```python
saya_crop_list[0].save("save_anim.gif",
                     save_all=True,
                     append_images=saya_crop_list[1:4],
                     loop=0,
                     duration=200)
```

```python
# 赤のみの単色化
saya_r = np.zeros((saya_org.height, saya_org.width, 3), dtype=np.uint8)
saya_r[:, :, 0] = saya_org_array[:, :, 0]

# 緑のみの単色化
saya_g = np.zeros((saya_org.height, saya_org.width, 3), dtype=np.uint8)
saya_g[:, :, 1] = saya_org_array[:, :, 1]

# 青のみの単色化
saya_b = np.zeros((saya_org.height, saya_org.width, 3), dtype=np.uint8)
saya_b[:, :, 2] = saya_org_array[:, :, 2]

# 単色化した画像を横に並べて表示
saya_monos = Image.new("RGB", (saya_org.width * 3, saya_org.height))
saya_monos.paste(Image.fromarray(saya_r), (0, 0))
saya_monos.paste(Image.fromarray(saya_g), (saya_org.width, 0))
saya_monos.paste(Image.fromarray(saya_b), (saya_org.width * 2, 0))
plt.imshow(saya_monos)
```

```python
# 青だけ抜いた（赤と緑のみの）画像に加工
saya_rg = np.zeros((saya_org.height, saya_org.width, 3), dtype=np.uint8)
saya_rg[:, :, 0] = saya_org_array[:, :, 0]
saya_rg[:, :, 1] = saya_org_array[:, :, 1]
plt.imshow(saya_rg)
```

```python
# 色の強さ 128 を基準とした二値化
saya_gr_array = np.array(saya_org.convert("L"))
saya_wb1 = saya_gr_array // 128
plt.imshow(saya_wb1, cmap="gray")
```

```python
# 色の強さ 160 を基準とした二値化
saya_wb2 = saya_gr_array // 160

# 色の強さ 192 を基準とした二値化
saya_wb3 = saya_gr_array // 192

# 二値化した画像を横に並べて表示
saya_wbs = Image.new("L", (saya_org.width * 3, saya_org.height))
saya_wbs.paste(Image.fromarray(saya_wb1), (0, 0))
saya_wbs.paste(Image.fromarray(saya_wb2), (saya_org.width, 0))
saya_wbs.paste(Image.fromarray(saya_wb3), (saya_org.width * 2, 0))
plt.imshow(saya_wbs, cmap="gray")
```

```python
saya_r_trimmed = np.copy(saya_r)

# 赤の色について、色の強さ 192 を基準とした二値化
saya_r_wb = saya_r_trimmed[:, :, 0] // 192

# 赤のみの画像データと二値化したデータのかけ算
saya_r_trimmed[:, :, 0] *= saya_r_wb

# 赤の強さ 192 以上のみのピクセルを残した画像データの表示
plt.imshow(saya_r_trimmed)
```

```python
# 土台
saya_base = Image.new("RGB", (saya_org.width, saya_org.height))

# マスキング用のデータ
saya_r_trimmed_img = Image.fromarray(saya_r_trimmed[:, :, 0]).convert("L")

# マスク処理を実行して結果を表示
saya_masked = Image.composite(saya_base, saya_org, saya_r_trimmed_img)
plt.imshow(saya_masked)
```
